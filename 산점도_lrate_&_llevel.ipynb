#  Train Data scatterplot(llrate & llevel)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files

# 파일 업로드
uploaded = files.upload()

# 업로드한 파일 이름 확인
file_name = list(uploaded.keys())[0]

# 데이터 불러오기
data = pd.read_csv(file_name)

# scatter plot 그리기
for var_x in ['lrate', 'llevel']:
    for var_y in ['lrate', 'llevel']:
        if var_x != var_y:  # 같은 변수를 x와 y에 동시에 그릴 필요 없음
            sns.scatterplot(x=data[var_x], y=data[var_y])
            plt.xlabel(var_x)
            plt.ylabel(var_y)
            plt.title(f'Scatter plot of {var_x} vs {var_y}')

            # y=x 선 추가
            min_val = min(data[var_x].min(), data[var_y].min())
            max_val = max(data[var_x].max(), data[var_y].max())
            plt.plot([min_val, max_val], [min_val, max_val], 'r-')  # 빨간색 실선

            plt.show()
```
