```python
import pandas as pd

df_train = pd.read_csv('combined_train_data.csv')
df_valid = pd.read_csv('combined_val_data.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# 각 leaktype의 개수를 계산
leaktype_counts = df_train['leaktype'].value_counts()

# 개수를 전체 개수로 나누어 비율을 계산
leaktype_ratios = leaktype_counts / leaktype_counts.sum()

# 색상 매핑 설정
colors = leaktype_ratios.index.map(lambda x: 'orange' if x == 'in' else ('orange' if x == 'out' else 'green'))

# 수평 막대 그래프 시각화
plt.figure(figsize=(10, 6))
leaktype_ratios.plot(kind='barh', color=colors)
plt.title('Ratio of Each Leak Type in df_train')
plt.xlabel('Ratio')
plt.ylabel('Leak Type')
plt.savefig('leaktype_ratios_train.png')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 각 leaktype의 개수를 계산
leaktype_counts = df_valid['leaktype'].value_counts()

# 개수를 전체 개수로 나누어 비율을 계산
leaktype_ratios = leaktype_counts / leaktype_counts.sum()

# 색상 매핑 설정
colors = leaktype_ratios.index.map(lambda x: 'orange' if x == 'in' else ('orange' if x == 'out' else 'green'))

# 수평 막대 그래프 시각화
plt.figure(figsize=(10, 6))
leaktype_ratios.plot(kind='barh', color=colors)
plt.title('Ratio of Each Leak Type in df_valid')
plt.xlabel('Ratio')
plt.ylabel('Leak Type')
plt.savefig('leaktype_ratios_valid.png')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터프레임의 길이 계산
lengths = {'df_train': len(df_train), 'df_valid': len(df_valid)}

# 길이 데이터프레임 생성
lengths_df = pd.DataFrame(list(lengths.items()), columns=['Dataset', 'Length'])

# 막대 그래프 시각화
plt.figure(figsize=(8, 6))
plt.bar(lengths_df['Dataset'], lengths_df['Length'], color=['dodgerblue', 'crimson'])
plt.title('Length of df_train vs df_valid')
plt.xlabel('Dataset')
plt.ylabel('Number of Records')

# 그래프를 파일로 저장 (예: 'dataset_lengths.png')
plt.savefig('dataset_lengths.png')

# 그래프를 화면에 표시
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))
sns.histplot(df_train['10HZ'], kde=False, bins=50)  # bin 수를 조정 (예: 30)
plt.title('Histogram of 10HZ in Training Data')
plt.xlabel('10HZ')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 처음 10개 열을 선택합니다.
columns_to_plot = df_train.columns[:10]

# 2열 5행의 그리드를 생성합니다.
fig, axes = plt.subplots(5, 2, figsize=(20, 20))

# 히스토그램을 그립니다.
for i, col in enumerate(columns_to_plot):
    ax = axes[i//2, i%2]  # 그리드 위치 설정
    sns.histplot(df_train[col], kde=False, bins=50, ax=ax)
    ax.set_title(f'Histogram of {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

# 레이아웃 조정
plt.tight_layout()

# 그래프를 파일로 저장 
plt.savefig('columns_1_10_hist.png')

# 그래프를 표시합니다.
plt.show()
```
