```python
## Training Dataset 불러오기


import pandas as pd
import glob
import os

# 디렉토리 경로 설정
directory_path = r"C:\Users\조여원\Desktop\상하수도 데이터(상수관로 누수 감지)\Training\누수감지데이터셋(Training)"

# 지정된 디렉토리 내의 모든 CSV 파일 경로 검색
file_paths = glob.glob(os.path.join(directory_path, "*.csv"))

# 데이터프레임 리스트 초기화
dataframes = []

# 각 파일을 읽어와 데이터프레임 리스트에 저장
for file_path in file_paths:
    df = pd.read_csv(file_path)
    dataframes.append(df)

# 데이터프레임 합치기
combined_df = pd.concat(dataframes, ignore_index=True)

# 결과 출력
print(combined_df)

# 필요한 경우 합쳐진 데이터프레임을 파일로 저장
combined_df.to_csv(r"C:\Users\조여원\Desktop\combined_data.csv", index=False)
```
