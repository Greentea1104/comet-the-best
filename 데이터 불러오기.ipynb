### 각각의 데이터셋 불러오기

```python
import pandas as pd
df1_train = pd.read_csv('1.옥외누수(out-training).csv')
df2_train = pd.read_csv('2.옥내누수(in-training).csv')
df3_train = pd.read_csv('3.기계.전기음(noise-training).csv')
df4_train = pd.read_csv('4.환경음(other-training).csv')
df5_train = pd.read_csv('5.정상음(normal-training).csv')

df1_valid = pd.read_csv('1.옥외누수(out-validation).csv')
df2_valid = pd.read_csv('2.옥내누수(in-validation).csv')
df3_valid = pd.read_csv('3.기계.전기음(noise-validation).csv')
df4_valid = pd.read_csv('4.환경음(other-validation).csv')
df5_valid = pd.read_csv('5.정상음(normal-validation).csv')
```

### df_train, df_valid 하나로 만들기

```python
 df_train = pd.concat([df1_train, df2_train, df3_train, df4_train, df5_train], ignore_index=True)
```

```python
 df_valid = pd.concat([df1_valid, df2_valid, df3_valid, df4_valid, df5_valid], ignore_index=True)
```
