```python
import pandas as pd
import os
import zipfile
```

```python

# 압축 파일 경로와 해제 경로 설정
train_zip_path = r"C:\Users\life0\OneDrive\바탕 화면\산업인공지능 실습\누수감지데이터셋(Training).zip"
val_zip_path = r"C:\Users\life0\OneDrive\바탕 화면\산업인공지능 실습\누수감지데이터셋(Validation).zip"
train_extract_path = r"C:\Users\life0\OneDrive\바탕 화면\산업인공지능 실습\train_data"
val_extract_path = r"C:\Users\life0\OneDrive\바탕 화면\산업인공지능 실습\val_data"

# 압축 파일 해제
with zipfile.ZipFile(train_zip_path, 'r') as zip_ref:
    zip_ref.extractall(train_extract_path)

with zipfile.ZipFile(val_zip_path, 'r') as zip_ref:
    zip_ref.extractall(val_extract_path)

# CSV 파일만 필터링
train_files = [f for f in os.listdir(train_extract_path) if f.endswith('.csv')]
val_files = [f for f in os.listdir(val_extract_path) if f.endswith('.csv')]

# 트레이닝 데이터 로드
train_dfs = [pd.read_csv(os.path.join(train_extract_path, file)) for file in train_files]

# 밸리데이션 데이터 로드
val_dfs = [pd.read_csv(os.path.join(val_extract_path, file)) for file in val_files]

# 데이터 결합
train_data = pd.concat(train_dfs, ignore_index=True)
val_data = pd.concat(val_dfs, ignore_index=True)

# 결합된 데이터 저장
train_data.to_csv(r'C:\Users\life0\Downloads\combined_train_data.csv', index=False)
val_data.to_csv(r'C:\Users\life0\Downloads\combined_val_data.csv', index=False)

print('트레이닝 데이터와 밸리데이션 데이터를 성공적으로 결합하고 저장했습니다.')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 결합된 데이터 로드
train_data = pd.read_csv(r'C:\Users\life0\Downloads\combined_train_data.csv')
val_data = pd.read_csv(r'C:\Users\life0\Downloads\combined_val_data.csv')
```

```python
# 데이터셋 설명 출력
print("트레이닝 데이터셋 설명:")
print(train_data.describe())
print("\n밸리데이션 데이터셋 설명:")
print(val_data.describe())
```

```python
# 결측치 확인
print("\n트레이닝 데이터셋 결측치 개수:")
print(train_data.isnull().sum())
print("\n밸리데이션 데이터셋 결측치 개수:")
print(val_data.isnull().sum())
```

```python
# 히스토그램 시각화
plt.figure(figsize=(14, 7))
sns.histplot(train_data.iloc[:, 0], kde=True)
plt.title('Histogram of First Column in Training Data')
plt.show()

plt.figure(figsize=(14, 7))
sns.histplot(val_data.iloc[:, 0], kde=True)
plt.title('Histogram of First Column in Validation Data')
plt.show()
```

```python
import pandas as pd  # 데이터프레임을 다루기 위한 라이브러리
import numpy as np   # 수학 및 통계 연산을 위한 라이브러리
import matplotlib.pyplot as plt  # 데이터 시각화를 위한 라이브러리
import seaborn as sns            # Matplotlib 기반의 추가적인 시각화 기능을 제공하는 라이브러리
```

```python

# 상관 행렬 시각화
correlation_matrix_train = train_data.corr()
correlation_matrix_val = val_data.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix_train, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix - Training Data')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix_val, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix - Validation Data')
plt.show()
```
